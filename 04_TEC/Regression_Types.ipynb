single variable regression, multiple variable regression, and polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Number of data points
num_data_points = 100

# Generate single variable regression data
x_single = np.linspace(0, 10, num_data_points)
y_single = 3 * x_single + 2 + np.random.randn(num_data_points) * 2

# Generate multiple variable regression data
x_multi = np.random.rand(num_data_points, 2) * 10
y_multi = 2 * x_multi[:, 0] - 3 * x_multi[:, 1] + 5 + np.random.randn(num_data_points) * 2

# Generate polynomial regression data
x_poly = np.linspace(-5, 5, num_data_points)
y_poly = 2 * x_poly**3 - 3 * x_poly**2 + 4 * x_poly + 5 + np.random.randn(num_data_points) * 10

# Plot the data
plt.figure(figsize=(15, 5))

# Single variable regression plot
plt.subplot(1, 3, 1)
plt.scatter(x_single, y_single, label='Data')
coeff_single = np.polyfit(x_single, y_single, 1)  # Fit a linear regression
y_single_fit = np.polyval(coeff_single, x_single)
plt.plot(x_single, y_single_fit, 'r', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Single Variable Regression')
plt.legend()

# Multiple variable regression plot
plt.subplot(1, 3, 2)
plt.scatter(x_multi[:, 0], y_multi, label='Data')
coeff_multi = np.polyfit(x_multi[:, 0], y_multi, 1)  # Fit a linear regression
y_multi_fit = np.polyval(coeff_multi, x_multi[:, 0])
plt.plot(x_multi[:, 0], y_multi_fit, 'r', label='Regression Line')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Multiple Variable Regression')
plt.legend()

# Polynomial regression plot
plt.subplot(1, 3, 3)
plt.scatter(x_poly, y_poly, label='Data')
coeff_poly = np.polyfit(x_poly, y_poly, 3)  # Fit a polynomial of degree 3
y_poly_fit = np.polyval(coeff_poly, x_poly)
plt.plot(x_poly, y_poly_fit, 'r', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()

plt.tight_layout()
plt.show()
```
