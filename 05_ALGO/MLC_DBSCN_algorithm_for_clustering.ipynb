# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
This is a clustering algorithm for data mining. It groups data points that are close to each other based on their density in the feature space.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standardize the features for better performance of DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Plot the clusters and noise points
plt.scatter(X_scaled[labels != -1, 0], X_scaled[labels != -1, 1], c=labels[labels != -1], cmap='viridis', label='Clusters')
plt.scatter(X_scaled[labels == -1, 0], X_scaled[labels == -1, 1], c='red', marker='x', label='Noise/Outliers')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title('DBSCAN Clustering on Iris Dataset')
plt.legend()
plt.show()
```

Another DBSCAN example with sinthetic data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate synthetic data with three blobs
X, y = make_blobs(n_samples=300, centers=[[1, 1], [-1, -1], [1, -1]], cluster_std=0.4, random_state=42)

# Standardize the features for better performance of DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Plot the clusters and noise points
plt.scatter(X_scaled[labels != -1, 0], X_scaled[labels != -1, 1], c=labels[labels != -1], cmap='viridis', label='Clusters')
plt.scatter(X_scaled[labels == -1, 0], X_scaled[labels == -1, 1], c='red', marker='x', label='Noise/Outliers')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title('DBSCAN Clustering on Synthetic Data')
plt.legend()
plt.show()
```

inference with iris dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features for better performance of DBSCAN
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Number of clusters, ignoring outliers (noise points)
num_clusters = len(np.unique(labels)) - 1

# Inference
for cluster_label in range(num_clusters):
    cluster_samples = X[labels == cluster_label]
    cluster_species = y[labels == cluster_label]
    species_counts = np.bincount(cluster_species)
    most_common_species = np.argmax(species_counts)

    print(f"Cluster {cluster_label + 1}:")
    print(f"Number of samples in cluster: {len(cluster_samples)}")
    print(f"Most common species: {most_common_species}\n")

# Plot the clusters and noise points
plt.scatter(X_scaled[labels != -1, 0], X_scaled[labels != -1, 1], c=labels[labels != -1], cmap='viridis', label='Clusters')
plt.scatter(X_scaled[labels == -1, 0], X_scaled[labels == -1, 1], c='red', marker='x', label='Noise/Outliers')
plt.xlabel('Sepal Length (Standardized)')
plt.ylabel('Sepal Width (Standardized)')
plt.title('DBSCAN Clustering on Iris Dataset')
plt.legend()
plt.show()
```
