## Predicting weight using linear regression

### Import the height-weight data csv

```python
import numpy
import pandas as pd

url='https://raw.githubusercontent.com/ShilpaSayuraML/Machine-Learning-in-7-Days/master/Section%202/height-weight.csv'
data = pd.read_csv(url, names = ['Height','Weight'])
```

```python
# We check the size of the data
data.info()
```

```python
# We check the first 10 rows
data[:10]
```

```python
#We import the plotting library 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
#We separate out the independent variable height into X 
#and dependent variable weight into y
X=data['Height'].to_numpy()
y=data['Weight'].to_numpy()
```

```python
data[:]
```

```python
X[0], y[0]
```

```python
X_50 = X[:50]
y_50 = y[:50]
plt.scatter(X_50, y_50, color='red', s=30) ## here s is size
```

```python
# Split the data into training/testing sets
X_train=X[:4500]
X_test=X[4500:]
# Split the targets into training/testing sets
y_train=y[:4500]
y_test=y[4500:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Modeify the data to input into sklearn
X_train=X_train.reshape(-1, 1)
X_test=X_test.reshape(-1, 1)
```

```python
X_train.shape
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
regr.predict([[70],[75]])
```

```python
# Plot outputs
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

#preditions
plt.axvline(x =70, ymin = 0.0, ymax = 240, color ='red')
plt.axhline(y=192.7, xmin=0, xmax=75, c='r')
#plt.xticks(())
#plt.yticks(())

plt.show()
```
