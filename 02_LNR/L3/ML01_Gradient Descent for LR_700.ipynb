#Gradient Descent for Simple Linear Regression

Gradient Descent minimizes some cost function by iteratively moving in the direction of steepest descent, updating the parameters along the way. gradient_descent-1.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# The dataset
X = np.array([[948],[1611],[830],[1453],[1656],[867],[861],[1129]])
y = np.array([[1307],[1734],[1176],[1601],[1654],[1240],[1196],[1435]])
```

```python
mu=np.mean(X)
sigma=np.std(X)
X=(X-mu)/sigma
print("mu: ",mu,"sigma: ",sigma)
print("Standarized Dataset: \n",X)
```

We standardized the input feature x.

```python
m = 0
b = 0
learning_rate = 0.1  
iterations = 100
n = len(X) 
cost_history = np.zeros(iterations)
```

We defined the initial parameters to 0, set our learning rate to 0.1, the iterations to 100, and n the length of the dataset, in “cost_history” we will store the cost after each iteration.

```python
for i in range(iterations): 
    ypred = m * X + b  
    dm = (1/n) * sum(X * (ypred - y))  
    db = (1/n) * sum(ypred - y)  
    m = m - learning_rate * dm  
    b = b - learning_rate * db  
    cost_history[i] = (1 / (2 * n)) * np.sum(np.square(ypred - y))
    
print (m, b, cost_history[i])
```

Here we implement the gradient descent algorithm, as you can see we used the derivate of the cost function with respect to each parameter and update the values after each iteration, finally printing the values for m, b, and the last cost.

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.set_ylabel('Cost')
ax.set_xlabel('Iterations')
ax.plot(range(iterations),cost_history,'b.')
```

As you can see in the plot, at some point iterating over doesn’t give us better results.
