* our model will produce an output for any given input. We can check these estimated outputs against the actual values we predicted. We call this difference between the actual value and the models estimate a **residual**.

* We can calculate the residual for every data point in our data set. These residuals will play a significant role in judging the usefulness of a model.

#Mean Squared Error (MSE) 
MSE is a procedure for measuring the average of the squares of the errors. it is the average squared difference between the actual values and what is estimated.


MNskFmGPKuQfMLdmpkT-X7-8w2cJXulP3683.png

* The purple dots are the actual points on the graph.

* The blue line is the prediction line. It passes through all the points fitting them in the best way. This line contains the predicted points.

The red line between each purple point and the prediction line are the errors. Each error has the distance from the point to its predicted point.

# Remember the y=Mx+B formula. M is the slope of the line and B is y-intercept of the line.

> We want to find M (slope) and B (y-intercept) that minimizes the squared error.

For each point, The y-coordinate is the purple dot. The y point is on the line we created. 
(y-y')^2  is the square of the result.

We sum all such results and divide by number ov points (n) to obtain the mean or MSE.

#Mean Absolute Error (MAE)
We calculate the residual for every data point, and take only the absolute value of each so that negative and positive residuals do not cancel out. We then take the average of all these residuals. 

BmBC8VW.jpg

A small MAE suggests our model is great at prediction. A large MAE suggests that your model may have trouble in certain areas. A MAE of 0 means that your model is a perfect predictor. But it may never happdn.



```python
#calculate MAE

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# input data
x = np.array([10, 20, 40, 80, 160])
y = np.array([76.6, 118.6, 200.8, 362.3, 648.9])

# sklearn linear regression
reg = LinearRegression().fit(x.reshape(- 1, 1), y)
y_pred = reg.predict(x.reshape(- 1, 1))

mse = mean_squared_error(y_true=y, y_pred=y_pred, squared=True)

print('sklearn intercept: ', reg.intercept_)
print('sklearn slope: ', reg.coef_[0])
print('sklearn MSE:', mse)
```
