## Linear Regression

#BMI

Body mass index (BMI) is a measure of body fat based on height and weight that applies to adult men and women.

```python
# importing libraries
#for versions checkout requirements.txt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing and reading data sets
url="https://raw.githubusercontent.com/purnasai/Linear_regression_with_blog/master/linear_regrerssion/bmi_and_life.csv"
data = pd.read_csv(url)
```

```python
#look at top 5 rows in data set
data.head()
```

```python
data[data['Country']=='Sri Lanka']
```

```python
#get the info
data.info()
```

from the above we can say that we have a total of 163 entries/rows, and 3 columns/features. 

```python
#delete/drop 'Country' variable.
data = data.drop(['Country'], axis = 1)
```

```python
data.describe()
```

```python
#get the shape
data.shape
```

it says that you have 163 rows and 2columns

```python
#reading into variables
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
```

## Split the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size = 20, random_state = 0)
```

## Model import

```python
#import linear regression
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#fitting the model
lr.fit(X_train,y_train)
```

```python
# Predicting the Test set results
y_pred = lr.predict(X_test)
```

## Visualisation

```python
#traininig set
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Life Expectancy vs BMI (Training set)')
plt.xlabel('Life Expectancy')
plt.ylabel('BMI')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Life Expectancy vs BMI (Test set)')
plt.xlabel('Life Expectancy')
plt.ylabel('BMI')
plt.show()
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, y_pred), 2))
```
