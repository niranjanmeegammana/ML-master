#Linear Regression 

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel, color='red')
plt.show()
```

```python
#predit future values
speed=myfunc(16)
print(speed)
```

```python
plt.scatter(x,y)
plt.plot(x, mymodel, color='red')
plt.plot(16, speed, marker = 'o', color='black')
plt.axhline(y = speed, color = 'orange', linestyle = '-')
plt.axvline(x = 16, color = 'orange', linestyle = '-')
```
