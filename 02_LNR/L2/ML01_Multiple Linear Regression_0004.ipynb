# Multiple Linear Regression

```python
# Import packages and classes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# Read the cars.csv file
url="https://raw.githubusercontent.com/ShilpaSayuraML/data-science-mastsr/master/data/cars.csv"
data = pd.read_csv(url)
data.head()
```

```python
# Print first 5 data
data.describe()
```

```python
# Print information of Dataset
data.info()
```

```python
# Divide the data into “Attributes” and “labels”
X = data[['Weight', 'Volume']]
y = data['CO2']
```

```python
# Split 80% of the data to the training set while 20% of the data to test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create a Linear Regression model and fit it
regressor =LinearRegression(fit_intercept=True)
regressor.fit(X_train,y_train)
```

```python
# Getting Results
print('Linear Model Coeff (m) =' , regressor.coef_)
print('Linear Model Coeff (b) =' , regressor.intercept_)
```

```python
# Predicting the data
y_predict=regressor.predict(X_test)
print(y_predict)
```

```python
# Prediction the CO2 emission of car using Weight and Volume Value of the car directly
print('---------[700,900]---------')
wg = 700
vol = 900
co2 = regressor.predict([[wg,vol]])
print(co2)
print('--------[1100,1500]----------')
wg = 1100
vol = 1500
co2 = regressor.predict([[wg,vol]])
print(co2)
print('--------[1500,2500]----------')
wg = 1500
vol = 2500
co2 = regressor.predict([[wg,vol]])
print(co2)
```
