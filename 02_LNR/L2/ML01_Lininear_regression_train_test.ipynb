# Train/Test
# How good is your model?
In Machine Learning we create models to predict the outcome of certain events

Train/Test is a method to measure the accuracy of your model.

WecIt split the data set into two sets: 
They are called training set and the testing set.

we use 80% of da5afor training, and 20% for testing.or split data 60%:40% into two sers.

We train the model using the training set.
We test the model using the testing set.

```python
#this data set has 100 customers, and their shopping habits.

import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100) # time in mins stayed in the shop by customers
y = numpy.random.normal(150, 40, 100) / x # purchase amount

y
```

```python
plt.scatter(x, y)
plt.show()
```

# Split Into Train/Test
The training set should be a random selection of 80% of the original data.

The testing set should be the remaining 20%.


```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

print(train_x)
```

```python
print(train_y)
```

```python
print(test_x[0], test_y[0])
```

```python
plt.scatter(train_x, train_y, alpha=0.5)
plt.scatter(test_x, test_y, color='r')
plt.show()
```

#Fit the Data Set
The best fit would be a polynomial regression, so let us draw a line of polynomial regression.



```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline), color='green')
plt.show()
```

#R-squared

```python
from sklearn.metrics import r2_score

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

#Bring in the Testing Set

Now we test the model with the testing data, to see if gives us the same result.

```python
r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

The result 0.809 shows that the model fits the testing set as well, and we are confident that we can use the model to predict future values.
